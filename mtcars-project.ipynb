#### Importing all the necessary libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

#### Let's start by loading dataset

```python
# loading the mtcars dataset into a pandas dataframe
mtcars_df = pd.read_csv('/Users/feruzkholov/Documents/mtcars.csv')

# printing the first five rows of the dataset to verify it loaded correctly
print(mtcars_df.head())
```

```python
# getting the summary statistics of the dataset
print(mtcars_df.describe())
```

```python
# checking for missing values
print(mtcars_df.isnull().sum())
```

```python
# checking for duplicated rows
print(mtcars_df.duplicated().sum())
```

```python
# getting the correlation matrix for the dataset
corr_matrix = mtcars_df.corr()
print(corr_matrix)
```

#### Data Visualisation

```python
# visualizing the correlation matrix with a heatmap
fig, ax = plt.subplots(figsize=(10,10)) 
sns.heatmap(corr_matrix, annot=True, linewidth=0.5, cmap='crest', vmin=-1, vmax=1,ax=ax)
plt.title('Correlation Matrix')
plt.show()
```

```python
# create a boxplot to compare the mpg variable across different numbers of cylinders
fig, ax = plt.subplots(figsize=(5,5)) 
sns.boxplot(data=mtcars_df, x="cyl", y="mpg",palette="Set3", hue='cyl', ax=ax)
plt.title('Comparison of "mpg" with "cylinders"')
plt.show()
```

```python
# create a scatterplot to visualize the relationship between horsepower and mpg
fig, ax = plt.subplots(figsize=(8,8)) 
sns.scatterplot(x='hp', y='mpg', data=mtcars_df,hue='gear',style='cyl', palette='deep',s=100, alpha=0.7)
plt.title('Relationship between "horsepowe" and "mp"')
plt.show()
```

```python
# create a density plot of the mpg variable
plt.figure(figsize=(10,10))
sns.kdeplot(data=mtcars_df, x=mtcars_df['wt'], hue="gear",bw_adjust=2,multiple="stack", common_norm=False, 
            palette="crest",alpha=.6, linewidth=0)
plt.title('Density plot of "mpg" ')
plt.show()
```

```python
# create a violin plot to compare the mpg variable across different numbers of cylinders
plt.figure(figsize=(10,10))
sns.violinplot(x='cyl', y='mpg', data=mtcars_df, 
               hue='gear',bw=1.5,inner="quartile", 
               palette='pastel', saturation=.8)
plt.title('Comparison of "mpg" with "cylinders"')
plt.show()
```

```python
# create a jointplot to visualize the relationship between horsepower and mpg
sns.jointplot(x='hp', y='mpg', data=mtcars_df, kind='reg',palette='muted')
plt.show()
```

```python
# create a lmplot to visualize the relationship between horsepower and mpg, stratified by number of cylinders
sns.lmplot(x='hp', y='mpg', data=mtcars_df, hue='cyl',col="am",  height=6)
plt.show()
```

```python
# create a rugplot to visualize the distribution of mpg values
plt.figure(figsize=(6,6))
sns.rugplot(mtcars_df['mpg'])
sns.scatterplot(data=mtcars_df, x="mpg", y="hp", hue='cyl', palette='Set1', s=100, alpha=.4)
sns.rugplot(data=mtcars_df, x="mpg", y="hp", height=.1, hue='cyl', lw=1,palette='Set1')
plt.title('Distribution of "mpg" values')
plt.show()
```

```python
# calculate the average mpg values for each combination of cylinders and gears
avg_mpg = mtcars_df.groupby(['cyl', 'gear'])['mpg'].mean().reset_index()
avg_mpg
```

```python
# plot the average mpg values using a barplot
sns.barplot(x='cyl', y='mpg', palette='Blues', data=avg_mpg,
            capsize=.1,errorbar="sd", linewidth=3,  alpha=0.4,errcolor='#34b4eb', edgecolor = '#34b4eb')
plt.title('Average "mpg" values')
plt.show()
```

```python
# create a heatmap to visualize the distribution of mpg values across different numbers of cylinders and gears
pivot_table = pd.pivot_table(mtcars_df, values='mpg', index=['cyl'], columns=['gear'])
ax = sns.heatmap(pivot_table,cmap='coolwarm',fmt='.1f', annot=True)
ax.set(xlabel="gear", ylabel="cyl")
ax.xaxis.tick_top()
plt.show()
```
